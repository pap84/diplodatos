```python
permits = pd.read_csv("AnalisisYCuracion/input/Building_Permits.csv",parse_dates=["Current Status Date","Filed Date","Issued Date","Completed Date"
                                                                   ,"First Construction Document Date","Permit Expiration Date"])
permits.dtypes
```

```python
permits.columns.str.extract(r'^(\w+)$')
```

```python
permits.columns = permits.columns.str.replace(' ','_')
permits.columns = permits.columns.str.replace('-','and')
permits.columns
```

```python
#Veamos si hay duplicados
permits[permits.duplicated()]
#Gracias a esto podemos ver que no hay registros duplicados.
```

```python
#veamos las caracteristicas de nuestras columnas
permits.describe(include="all")
```

31 de las 43 columnas tienen missing values, veremos cuales tienen un porcentaje demasiado alto.

```python
#Ahora veremos que hacemos con los valores perdidos
missing_values_count = 100*permits.isnull().sum()/permits.shape[0]
delcol = missing_values_count[missing_values_count > 85].sort_values(ascending=False)
delcol
```

Las columnas sin valor en la mayoría (>85%) de sus filas, son eliminadas porque no permitirán análisis.

```python
permits = permits.drop(delcol.index.tolist(), axis=1)
```

```python
permits.shape
```

## Convertir variables categóricas

```python
from sklearn import preprocessing
```

```python
permits.describe(include=['O'])
```

```python
columns = ["Permit_Type_Definition","Current_Status","Permit_Type_definition"]
for column in columns:
    try:
        permits[column] = permits[column].str.replace(' ','_')
    except:
        continue
    le = preprocessing.LabelEncoder()
    le.fit(permits[column])
    dict(zip(le.classes_,le.transform(le.classes_)))
    permits[column] = le.transform(permits[column])
    
permits.head()


    
```
