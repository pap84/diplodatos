```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

lenage = pd.read_csv("input/lenageinf.csv")
wtage =pd.read_csv("input/wtageinf.csv")
```

```python
male_lenage=lenage.where(lenage.Sex == '1')
male_lenage=male_lenage.dropna()
female_lenage = lenage.where(lenage.Sex == '2')
female_lenage=female_lenage.dropna()
```

Curvas de Altura vs Edad para los diferentes percentiles
===============================================

```python
lenage.dtypes
```

```python
pes = ["P3","P5","P10","P25","P50","P75","P90","P95","P97"]
fig,axes =  plt.subplots(9,1,figsize=(25,90))
i = 0
for p in pes:
    
    axes[i].plot(male_lenage.Agemos.apply(float),male_lenage[p].apply(float),'-',label="Hombre",)
    axes[i].plot(female_lenage.Agemos.apply(float),female_lenage[p].apply(float),'-',label="Mujer")
    axes[i].set_xlabel("Edad en meses")
    axes[i].set_ylabel("Altura")
    axes[i].set_title("Curvas de Altura correspondiente al percentil "+p)
    axes[i].legend()
    i = i+1
plt.tight_layout()
```
